```python
#importing necessary libraries
import pandas as pd
import numpy as np
```

```python
#importing dataset
df=pd.read_csv('/content/heart.csv')
df.head()
```

```python
df.info()
```

```python
df['target'].value_counts()
```

```python
#extracting indipendent and dependent columns
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
#splitting data into training and testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier(100)
clf.fit(x_train,y_train)
```

```python
clf.score(x_train,y_train),clf.score(x_test,y_test)
```

**model is overfitting so we applying Fearure importance for best Optimal model**

#Feature Importance

```python
clf.feature_importances_
```

```python
feature_importance = pd.DataFrame({'importance': clf.feature_importances_}, index= x.columns).sort_values('importance')
```

```python
feature_importance
```

```python
feature_importance[feature_importance['importance']>0.05].index
```

**we are selecting features only which are important**

```python
to_keep=list(feature_importance[feature_importance['importance']>0.05].index)
```

```python
x=x[to_keep]
```

```python
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y,
                                                    test_size = 0.2,
                                                    random_state=0)
```

```python
clf.fit(x_train,y_train)
```

```python

from sklearn.model_selection import RandomizedSearchCV
```

```python
param_grid = {
    'n_estimators': [15,5],
    'max_depth': [None, 10, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2', None],
    'bootstrap': [True, False]
}
```

```python
rd_search=RandomizedSearchCV(clf, param_distributions = param_grid, cv = 5, n_iter = 10)
```

```python
rd_search.fit(x_train,y_train)
```

```python
rd_search.score(x_train,y_train),rd_search.score(x_test,y_test)
```
